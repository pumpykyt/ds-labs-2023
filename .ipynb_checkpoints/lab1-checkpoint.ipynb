```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Дані
Ініціалізація масиву даних, зчитування датасетів, конкатинація даних в один масив

```python
arr = []
for i in range(1, 32):
    day = pd.read_csv('./data/dds_ch2_nyt/nyt' + str(i) + '.csv')
    day['Day'] = i
    arr.append(day)
d = pd.concat(arr)
df = d.copy()
bins = [-1, 0, 18, 25, 35, 45, 55, 65, float('inf')]  # Age group boundaries
labels = ['N', '<18', '18-24', '25-34', '35-44', '45-54', '55-64', '65+']  # Labels for age groups

# Create the 'age_group' column
df['age_group'] = pd.cut(df['Age'], bins=bins, labels=labels)

# separate signed users
df_s = df[df['Signed_In'] == 1]
df_u = df[df['Signed_In'] == 0]

df_s = df_s.drop('Signed_In', axis=1)

print(df)
print(df_s)
```

```python
age_group_data = df.groupby('age_group').agg({'Impressions': 'sum', 'Clicks': 'sum'}).reset_index()
# Calculate CTR for each age group
age_group_data['CTR'] = age_group_data['Clicks'] / age_group_data['Impressions']

plt.figure(figsize=(10, 6))
sns.barplot(x='age_group', y='Impressions', data=age_group_data)
plt.title('Impressions by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Impressions')

plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x='age_group', y='CTR', data=age_group_data)
plt.title('CTR by Age Group')
plt.xlabel('Age Group')
plt.ylabel('CTR')

plt.show()
```

###
з результатів видно, що найбільше показів мають люди з вікової групи 35-44  
можна припустити, що ресурс орієнтований саме на цю вікову групу, так як це найбільш впливова на роботоспроможна група суспільства (в середньому)  
тому інформаційним джерелам вигідно подавати необхідну інформацію саме цій категорії людей  

щодо переходів на сторінки, то найбільшу кількість мають вікові категорії до 18 та після 65. Це можна пов'язати з тим, що ці люди мають більше часу на детальний перегляд контенту ніж ті що працюють

```python
days_signed = df_s.groupby(['age_group', 'Gender']).agg({'Impressions': 'sum', 'Clicks': 'sum'}).reset_index()
days_signed['CTR'] = days_signed['Clicks'] / days_signed['Impressions']

plt.figure(figsize=(12, 6))
sns.barplot(x='age_group', y='Impressions', hue='Gender', data=days_signed, palette =['#fcc','#55f'])
plt.title('Impressions by Age Group and Gender')
plt.xlabel('Age Group')
plt.ylabel('Number of Impressions')
plt.legend(title='Gender (female/male)', loc='upper right')

plt.show()

clicks_amount = df_s.groupby(['age_group', 'Gender']).agg({'Impressions': 'sum', 'Clicks': 'sum'}).reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='age_group', y='Clicks', hue='Gender', data=clicks_amount, palette =['#fcc','#55f'])
plt.title('Clicks by age group (Signed)')
plt.xlabel('Age Group')
plt.ylabel('Clicks')
plt.legend(title='Gender (female\male)', loc='upper right')

plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x='age_group', y='CTR', hue='Gender', data=days_signed, palette =['#fcc','#55f'])
plt.title('CTR by Age Group and Gender')
plt.xlabel('Age Group')
plt.ylabel('CTR')
plt.legend(title='Gender (female/male)', loc='upper right')

plt.show()

# plt.figure(figsize=(12, 6))
# sns.lineplot(x='Day', y='CTR', hue='Signed_In', data=gender_amount)
# plt.title('CTR per day (Signed)')
# plt.xlabel('Day of month')
# plt.ylabel('CTR')
# plt.legend(title='Signed In', loc='upper right')

# plt.show()
```

###
цікаві дані щодо гендерів відвідувачів  
у вихідні кількість переглядів майже однакова, але різниця не настільки велика щоб робити точні висновки  

щодо переходів на ресурси, то тут жінки трохи переважають, можна прив'язати це до того, що в жінок є більше можливостей чи часу щоб детально переглядати сторінки, або зробити висновок, що контент зацікавлює жінок трохи більшою мірою

```python
def getLabel(i):
    if i == 0:
        return 'Female'
    return 'Male'

grouped_metrics = df.groupby(['age_group', 'Gender']).agg({'Impressions': 'sum', 'Clicks': 'sum', 'Signed_In': 'sum'})

grouped_metrics['CTR'] = grouped_metrics['Clicks'] / grouped_metrics['Impressions']

# create separate columns for Impressions, Clicks, and CTR by Gender
pivot_df = grouped_metrics.pivot_table(index='age_group', columns='Gender', values=['Impressions', 'Clicks', 'CTR', 'Signed_In'])

# rename the columns
pivot_df.columns = [f'{col[0]}_{getLabel(col[1])}' for col in pivot_df.columns]

# reset the index to make age_group a regular column
pivot_df.reset_index(inplace=True)

pivot_df
```
